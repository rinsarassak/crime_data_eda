# Introduction

In 2020, crime data played a pivotal role in analyzing patterns of criminal activity and understanding the social dynamics of various communities. The data, which includes detailed records such as the date a crime was reported, the date it occurred, geographic location, and specific crime codes, provides a comprehensive overview of criminal incidents within various jurisdictions. Each entry not only identifies the type of crime but also includes additional factors such as the age, sex, and descent of the victim, as well as the nature of the premises and whether any weapons were involved.The data serves as a key tool for law enforcement agencies, policymakers, and researchers, helping to track criminal trends and improve public safety measures. It provides valuable insights into the frequency and distribution of different crime types across regions, highlighting areas where additional resources or interventions may be needed. Moreover, the status of each case—whether resolved, ongoing, or unsolved—further contributes to understanding the effectiveness of law enforcement and the challenges faced in solving cases.By analyzing such data, it becomes possible to identify potential correlations between demographic factors, crime trends, and geographic regions, thus facilitating more informed decisions in crime prevention strategies. Ultimately, this information is crucial in shaping public policy, enhancing community safety, and fostering a deeper understanding of the underlying causes of crime within society.





- DR_NO: Unique crime report number
- DATE_RPTD: Date the crime was reported
- DATE_OCC: Date the crime occurred
- AREA: Area number where the crime occurred
- AREA_NAME: Name of the area where the crime occurred
- RPT_DIST_NO: Reporting district number
- PART_1-2: Part 1 or Part 2 crime (Part 1 crimes are more serious)
- CRM_CD: Crime code
- CRM_CD_DESC: Description of the crime
- VICT_AGE: Age of the victim
- VICT_SEX: Sex of the victim
- VICT_DESCENT: Descent of the victim
- PREMIS_CD: Premise code (type of location)
- PREMIS_DESC: Description of the premise
- WEAPON_USED_CD: Weapon used code
- WEAPON_DESC: Description of the weapon used
- STATUS: Status of the case (Open or Closed)
- STATUS_DESC: Description of the status
- LOCATION: Location of the crime

#### 21. What are the peak times for crimes to occur 

```python
df['Date_rptd'] = pd.to_datetime(df['Date_rptd'])
df['Hour'] = df['Date_rptd'].dt.hour
df['Day_of_Week'] = df['Date_rptd'].dt.dayofweek
crime_counts = pd.pivot_table(df, index='Day_of_Week', columns='Hour', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(crime_counts, cmap='Blues', annot=True, fmt='d')
plt.title('Peak Times for Crimes to Occur')
plt.xlabel('Hour')
plt.ylabel('Day of the Week')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### 22. Which areas have the highest and lowest number of reported crimes 

```python
crime_counts = df['Area_name'].value_counts()
highest_crime_areas = crime_counts.nlargest(5)
lowest_crime_areas = crime_counts.nsmallest(5)
# Create a bar chart for the top 5 areas with the highest number of reported crimes
plt.figure(figsize=(10, 6))
highest_crime_areas.plot(kind='bar')
plt.title('Top 5 Areas with the Highest Number of Reported Crimes')
plt.xlabel('Area Name')
plt.ylabel('Number of Reported Crimes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create a bar chart for the bottom 5 areas with the lowest number of reported crimes
plt.figure(figsize=(10, 6))
lowest_crime_areas.plot(kind='bar')
plt.title('Bottom 5 Areas with the Lowest Number of Reported Crimes')
plt.xlabel('Area Name')
plt.ylabel('Number of Reported Crimes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### 23. Do certain streets have higher crime occurrences

```python
street_crime_counts = df.groupby('Cross_street')['Crm_cd'].count().reset_index(name='Count')
top_10_streets = street_crime_counts.nlargest(10, 'Count')
plt.figure(figsize=(10, 6))
plt.scatter(top_10_streets['Cross_street'], top_10_streets['Count'], s=top_10_streets['Count']*10, alpha=0.5)
plt.title('Top 10 Streets with Highest Crime Occurrences')
plt.xlabel('Cross_street')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
